$x$についての2次方程式 $ax^2+bx +c=0$の解は

$$
x = \frac{-b - \sqrt{b^2-4ac}}{2a}, \frac{-b + \sqrt{b^2-4ac}}{2a}
$$

により計算できる。$a, b, c$を入力し2次方程式の解を出力するプログラムを書け。平方根は`math.sqrt`で計算できる。

```python
# import math
a = 1
b = 2
c = -3
print((-b - math.sqrt(b**2 - 4 * a * c))/(2 * a))
print((-b + math.sqrt(b**2 - 4 * a * c))/(2 * a))
```

### Black-Scholes式
ヨーロピアンコールオプションの価格$C$は

$$
C = S N(d_1) - K e^{-rT} N(d_2)\\
d_1 = \frac{\ln(S/K) + (r +\sigma^2/2)T}{\sigma \sqrt{T}}\\
d_2 = d_1 - \sigma \sqrt{T}
$$

によって計算できる。
ここで、$S$は現在の原資産価格、$K$は権利行使価格、$T$は満期までの時間、$r$は無リスク金利、$\sigma$は原資産のボラティリティ、$N()$は標準正規分布の分布関数を表す。

標準正規分布の分布関数は `from scipy.stats import norm`としてから、`norm.cdf`を使えばよい。

$S = 100, K = 99, T = 1, r = 0.01, \sigma = 0.2$ のとき $C$ は約 $8.91$ となる。


```python
from scipy.stats import norm
import math
S = 100
K = 99
T = 1
r = 0.01
sigma = 0.2
d1 = (math.log(S / K) + (r + 0.5 * sigma**2) * T)/(sigma * math.sqrt(T))
d2 = d1 - sigma * math.sqrt(T)
C = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
print(C)
```
