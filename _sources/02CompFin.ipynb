## print関数



`print(x)`とすれば、変数`x`の値が表示された。計算結果の表示に使うのはもちろんのこと、プログラムの誤りを見つけるために適宜変数を`print`関数を使って表示させることも重要。

### format関数、f文字列 

文字列の中に数値を埋め込むには`format()`関数（メソッド）を用いる。


````python
a, b = 1, 2
print('{} plus {} equals {}'.format(a, b, a + b))
````

f文字列を使って、つぎのように書くこともできる。

````python
a, b = 1, 2
print(f'{a} plus {b} equals {a+b}')
````

`format()`関数、f文字列により、表示する文字列の書式を設定することもできる。例えば、小数点以下の桁数を指定するには `.(桁数)f`とする。表示の形式を指定するだけで、変数の中身が変わるわけではない。

````python
import math 
c = 5
d = math.sqrt(c)
print(d)
print('{:.3f}'.format(d))
````

f文字列ではつぎのようになる。

````python
import math 
c = 5
d = math.sqrt(c)
print(d)
print(f'{d:.3f}')
````

半径$r$を入力し、円の面積$\pi r^2$を出力するプログラムを作成せよ。ただし、出力は**小数点以下第3位まで**を表示せよ。

````python
import math
r = 2
S = math.pi * r ** 2
print(f'{S:.3f}')
````

## 条件分岐

`if`文により条件分岐（条件によって実行するプログラムを変更する）を行うことができる。以下が基本構文。
```
if 条件式:
  条件が成立したときの処理（ifブロック)
else:
  条件が成立しなかったときの処理(elseブロック)
```
ifブロック、elseブロックは**インデント**（字下げ、行頭に空白文字をいくつかを入れること）によって指定する。

Pythonではインデントとして半角スペース4個を使うことが多い。Google ColabではTabキーを押すと、半角スペース2個が入力される（設定により変更することもできる）。

````python
a, b = 3, 2
if a > b:
  print(a)
else:
  print(b)
````

条件式を書く際には比較演算子を使う。


|演算子|意味|
|--|--|
|a == b|aとbは等しい|
|a != b|aとbは等しくない|
|a > b|aはbより大きい|
|a < b|aはbより小さい|
|a >= b|aはb以上|
|a <= b|aはb以下|



````python
# x / y の計算
x = 1
y = 0
if y != 0:
  print(x / y)
else:
  print('NG')
````

## 繰り返し
`for`文または`while`文により、繰り返しをプログラムすることができる。ここでは、とりあえず`for`文だけ紹介する。つぎが基本構文になる。
```
for 変数 in リストなど:
  実行文
```


````python
numbers = [1, 2, 3, 5, 8, 13]
for i in numbers:
  print(i)
````

リスト`numbers`に含まれる数の和を`for`文を使って計算する。`a += b`は
`a = a + b`、すなわち、`a`に`a+b`を代入するという意味になる。これを複合代入演算子と呼び、`-=`, `*=`, `/=`なども使うことができる。

````python
ans = 0
for j in numbers:
  ans += j

print(ans)
````

リスト`numbers`に含まれる数の積を計算せよ。

````python
ans = 1
for j in numbers:
  ans *= j

print(ans)
````

````python
for k in range(3):
  print('Hello')
````

````python
for k in range(4):
  print(k)
````

````python
ans = 0
for l in range(1, 101):
  ans += l
print(ans)
````

## モンテカルロ法
Black-Scholesモデルの仮定の下では、$T$時点における株価$S(T)$は（リスク中立確率の下で）
```{math}
S(T) = S_0 \exp\left\{\left(r - \frac{\sigma^2}{2}\right)T + \sigma \sqrt{T} z\right\}
```
に従う。ここで、$S_0$は時点0（現時点）の株価、$\sigma$はボラティリティ、$r$は無リスク金利、$z$は標準正規分布（確率変数）。

このとき、権利行使価格$K$のヨーロピアンコールオプションの価格は
```{math}
e^{-rT}E[(S(T) - K)^+]  \tag{1} 
```
により計算できる。ここで、$(x)^+ = \max\{x, 0\}$である。
$(1)$は解析的に計算できて、いわゆるBlack-Scholes公式が導かれる。

ここではあえて、$(1)$をモンテカルロ法によって計算をしてみる。
$z_1, z_2, \dots, z_N$を標準正規分布からの標本とすると、$(1)$は
```{math}
\frac{1}{N} e^{-rT}\sum_{i=1}^N (S_i(T) - K)^+
```
```{math}
S_i(T) = S_0 \exp\left\{\left(r - \frac{\sigma^2}{2}\right)T + \sigma \sqrt{T} z_i\right\}
```
により近似できる。

具体的には以下のようにする。
$i=1,2,\dots,N$について以下の1-4の計算をする。Nは生成する標本の数。
1. 正規乱数$z_i$を発生させる．
2. $S_i(T) = S_0\exp\left\{\left(r - \frac12\sigma^2\right)T + \sigma
		       \sqrt{T}z_i\right\}$ を計算。
3. $V_i = (S^i(T) - K)^+$を計算。$i < N$のときは1に戻る。
4. 最後に，$\bar{V} = e^{-rT}(V_1 + V_2 + \dots + V_N)/N$を計算．

以下のプログラムでは、標準正規分布に従う乱数を `import random`として`random.gauss(0, 1)`により生成している。

Black-Scholes式によれば$S_0 = 100, K = 99, T = 1, r = 0.01, \sigma = 0.2$ のとき $C$ は約 $8.91$ となる。


````python
import math
import random

N = 10000 #生成する標本の数

S = 100
K = 99
T = 1
r = 0.01
sigma = 0.2

value = 0  #ペイオフを足し合わせる変数
for i in range(N):
  z = random.gauss(0, 1)
  price = S * math.exp((r - 0.5 * sigma ** 2)* T + sigma * math.sqrt(T) * z)
  if price > K:
    payoff = price - K
  else:
    payoff = 0
  value += payoff

print('{:.4f}'.format(math.exp(-r * T) * value / N))
````

実際には`if`文を使ってペイオフを求めるのではなく、`max()`関数を使った方がよい。`max(a, b)`とすることで、`a, b`のうち大きい方の値を得ることができる。

````python
import math
import random

N = 10000 #生成する標本の数

S = 100
K = 99
T = 1
r = 0.01
sigma = 0.2

value = 0  #ペイオフを足し合わせる変数
for i in range(N):
  z = random.gauss(0, 1)
  price = S * math.exp((r - 0.5 * sigma ** 2)* T + sigma * math.sqrt(T) * z)
  value += max(price - K, 0)

print('{:.4f}'.format(math.exp(-r * T) * value / N))
````

**練習問題** 
$z$を標準正規分布に従う確率変数とすると
```{math}
E[e^{kz}] = e^{\frac12k^2}  \tag{2}
```
が成り立つことが知られている。ここで、$k$は任意の実数である。
適当な$k$を与えて、この式の左辺をモンテカルロ法により計算し、右辺を計算した値と比較してみよ。モンテカルロ法の標本数$N$がどの程度の値であれば、(2)がほぼ成立するか調べよ。

````python
import math
import random
k = 1
N = 10000
value2 = 0
for i in range(N):
  z = random.gauss(0, 1)
  value2 += math.exp(k * z)

print('Monte Carlo: {:.3f}'.format(value2/N))
print('Analytical solution: {:.3f}'.format(math.exp(0.5 * k**2)))
````
